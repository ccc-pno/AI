```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

df_train = pd.DataFrame({'key1':['a', 'a', 'b', 'b', 'a'],
                                    'key2':['one', 'two', 'one', 'two', 'one'],
                                    'data1':[0,1,0,1,0]
                                    })
df_test = pd.DataFrame({'key1':['a','c','d','e'],
                                    'data1':[1,3,4,5]
                                    })
df_test2 = pd.DataFrame({'key1':['a','f','g','e'],
                                    'data1':[1,3,4,5]
                                    })
```

```python
# on 默认值是相同的列名
# how 默认inner(内连接)：保留两个数据集中整行完全一样的样本，这些样本的其他特征都会保留
# how：outer、left、right、inner
data = pd.merge(df_train, df_test, on=['key1'], how='left')
data
```

```python
data = pd.merge(df_test, df_test2, on=['key1'], how='inner')
data
```

```python
# concat 上下合并，列取两者所有的,  不存在的值为NaN

data = pd.concat((df_train, df_test),sort=False).reset_index(drop=True)
data
```
