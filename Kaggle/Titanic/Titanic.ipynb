```python
# 修补年龄，更具姓名判断平均年龄填充
Mr_avg_age = 0.
Mrs_avg_age = 0.
Miss_avg_age = 0.
Don_avg_age = 0.
Master_avg_age = 0.
Dr_avg_age = 0.
Ms_avg_age = 0.

Mr_count = 0.
Mrs_count = 0.
Miss_count = 0.
Don_count = 0.
Master_count = 0.
Dr_count = 0.
Ms_count = 0.

Don_names = []
Other_names = []

Nan_id = []

for id in range(csv.shape[0]):
    people = csv.values[id]
    if np.isnan(people[5]):
        Nan_id.append(id)
        continue
    if 'Mr.' in people[3]:
        Mr_avg_age += people[5]
        Mr_count += 1.
    elif 'Mrs.' in people[3]:
        Mrs_avg_age += people[5]
        Mrs_count += 1.
    elif 'Miss.' in people[3]:
        Miss_avg_age += people[5]
        Miss_count += 1.
    elif 'Don.' in people[3]:
        Don_avg_age += people[5]
        Don_count += 1.
    elif 'Master.' in people[3]:
        Master_avg_age += people[5]
        Master_count += 1.
    elif 'Dr.' in people[3]:
        Dr_avg_age += people[5]
        Dr_count += 1.
    if 'Ms.' in people[3]:
        Ms_avg_age += people[5]
        Ms_count += 1.
    else:
        Other_names.append(people[3])

Mr_avg_age /= Mr_count
Mrs_avg_age /= Mrs_count
Miss_avg_age /= Miss_count
Don_avg_age /= Don_count
Master_avg_age /= Master_count
Dr_avg_age /= Dr_count
Ms_avg_age /= Ms_count

for id in Nan_id:
    if 'Mr.' in csv['Name'][id]:
        csv['Age'][id] = Mr_avg_age
    elif 'Mrs.' in csv['Name'][id]:
        csv['Age'][id] = Mrs_avg_age
    elif 'Miss.' in csv['Name'][id]:
        csv['Age'][id] = Miss_avg_age
    elif 'Master.' in csv['Name'][id]:
        csv['Age'][id] = Master_avg_age
    elif 'Don.' in csv['Name'][id]:
        csv['Age'][id] = Don_avg_age
    elif 'Dr.' in csv['Name'][id]:
        csv['Age'][id] = Dr_avg_age
    elif 'Ms.' in csv['Name'][id]:
        csv['Age'][id] = Ms_avg_age
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA

def PolynomialLogisticRegression(degree=2, n_components=0.9, C=0.1, penalty='l2', solver='liblinear'):
    return Pipeline([
            ("poly", PolynomialFeatures(degree=degree)),
            ("std_scalert", StandardScaler()),
            ("pca", PCA(n_components=n_components)),
            ("log_reg", LogisticRegression(C=C,penalty=penalty,solver=solver))
    ] )
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = [
    {
        'poly__degree': [i for i in range(1,8)],
        'log_reg__C': [i for i in np.arange(0.1,0.8,0.1)],
        'log_reg__penalty':['l1','l2'],
        'pca__n_components':[0.9, 0.95]
    }
]

poly_logistic_reg2 = PolynomialLogisticRegression()
grid_search = GridSearchCV(poly_logistic_reg2, param_grid, verbose=2, cv=3) # cv是交叉验证分成多少分
grid_search.fit(X_train, y_train)
```
