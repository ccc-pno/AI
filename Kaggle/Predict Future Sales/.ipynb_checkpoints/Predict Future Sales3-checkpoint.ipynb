##### 您将获得每日历史销售数据。任务是预测测试集在每个商店中出售的产品总数。请注意，商店和产品的清单每月都会略有变化。创建可以处理此类情况的可靠模型是挑战的一部分。


##### 档案说明：
##### sales_train.csv-训练集。2013年1月至2015年10月的每日历史数据。
##### test.csv-测试集。您需要预测这些商店和产品在2015年11月的销售额。
##### sample_submission.csv-格式正确的示例提交文件。
##### items.csv-有关项目/产品的补充信息。
##### item_categories.csv-   有关项目类别的补充信息。
##### shop.csv-有关商店的补充信息。


##### 资料栏位：
##### ID-  代表测试集中的（商店，商品）元组的ID
##### shop_id-商店的唯一标识符
##### item_id-产品的唯一标识符
##### item_category_id-项目类别的唯一标识符
##### item_cnt_day-销售的产品数量。您正在预测该指标的每月金额
##### item_price-商品的当前价格
##### 日期  -格式为dd / mm / yyyy的日期
##### date_block_num-连续的月份号，为方便起见。2013年1月为0,2013年2月为1，...，2015年10月为33
##### item_name-  项目名称
##### shop_name-商店名称
##### item_category_name-项目类别名称

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from itertools import product
```

```python
df_test = pd.read_csv('test.csv').copy()
df_train = pd.read_csv('sales_train.csv').copy()
items = pd.read_csv('items.csv').copy()
# item_categories = pd.read_csv('item_categories.csv').copy()
# shops = pd.read_csv('shops.csv').copy()
# sample_submission = pd.read_csv('sample_submission.csv').copy()
```

# 观察数据

```python
plt.figure(figsize=(10,4))
sns.boxplot(x = df_train['item_cnt_day'])
plt.figure(figsize=(10,4))
sns.boxplot(x = df_train['item_price'])
```

```python
# 删除异常点，防止过拟合
df_train = df_train.drop(df_train[df_train['item_price']<0].index | df_train[df_train['item_price']>=100000].index)
df_train = df_train.drop(df_train[df_train['item_cnt_day']<0].index | df_train[df_train['item_cnt_day']>=100000].index)
```

# 特征工程

```python
# 增加三列：每商店每商品销售总量、商店后六个月是否关闭、item类型id
# 修改一列：价格改为平均价格
# 删date、id
```

```python
df_train
```

### 每商店每商品销售总量

```python
sales_by_itemshop = df_train.pivot_table(index=['item_id','shop_id'], columns=['date_block_num'], 
                                                values=['item_cnt_day'], aggfunc=[np.sum], 
                                                 fill_value=0).reset_index()
sales_by_itemshop.columns = sales_by_itemshop.columns.droplevel().droplevel().map(str)
sales_by_itemshop.columns.values[0] = 'item_id'
sales_by_itemshop.columns.values[1] = 'shop_id'
sales_by_itemshop = sales_by_itemshop.reset_index(drop=True).rename_axis(None, axis=1)
sales_by_itemshop
```

```python
# 记录一下最后一个月的销售量作为y值
sales_lastmonth_by_itemshop = pd.DataFrame({'item_id' : sales_by_itemshop['item_id'],
                                        'shop_id': sales_by_itemshop['shop_id'],
                                        'sales_last_month' : sales_by_itemshop.loc[:,'33']
                                    })
sales_lastmonth_by_itemshop
```

```python
sales_sum = pd.DataFrame({'item_id' : sales_by_itemshop['item_id'],
                                        'shop_id': sales_by_itemshop['shop_id'],
                                        'sales_sum' : sales_by_itemshop.loc[:,'0':].sum(axis=1)
                                    })
sales_sum
```

### 商店后六个月是否关闭

```python
zero_by_shop = df_train.pivot_table(index=['shop_id'], columns=['date_block_num'], values=['item_cnt_day'],
                                                   aggfunc=[np.sum], fill_value=0).reset_index()
zero_by_shop.columns = zero_by_shop.columns.droplevel().droplevel().map(str)
zero_by_shop.columns.values[0] = 'shop_id'
zero_by_shop = zero_by_shop.reset_index(drop=True).rename_axis(None, axis=1)
zero_by_shop
```

```python
close_shop = pd.DataFrame({'shop_id': zero_by_shop['shop_id'],
                                        'close_shop' : (zero_by_shop.loc[:,'28':].sum(axis=1)==0).astype(int)
                                      })
close_shop
```

### 平均价格

```python
mean_price_by_itemshop = df_train[['item_id','shop_id','item_price']].groupby(['item_id','shop_id'])['item_price']
mean_price = pd.DataFrame(mean_price_by_itemshop.mean()).reset_index()
mean_price
```

```python
df_train_size = df_train.shape[0]
df_test_size  = df_test.shape[0]
data = pd.DataFrame()
data = pd.concat((df_train, df_test),sort=False).reset_index(drop=True)
data = pd.merge(data, items, how='left')
data = pd.merge(data, sales_sum, how='left')
data = pd.merge(data, close_shop, how='left')
data = pd.merge(data, mean_price, how='left')

data = data.drop(['date','ID','item_name','item_price','item_cnt_day'],axis=1)
data['date_block_num'].fillna(34, inplace=True)
data['sales_sum'].fillna(0, inplace=True)


data
```

```python
data.isnull().sum()
```

```python
data = data.drop(['date_block_num'],axis=1)
data = pd.merge(data, sales_lastmonth_by_itemshop, how='left')

data
```

```python
data.isnull().sum()
```

```python
data.head(5)
```

```python
shop_id = data['shop_id'].drop_duplicates().reset_index().drop(['index'],axis=1)
item_id  = data['item_id'].drop_duplicates().reset_index().drop(['index'],axis=1)
data3 = pd.DataFrame(product(shop_id.values.reshape(1,-1)[0], item_id.values.reshape(1,-1)[0]), columns=['shop_id','item_id'])
data3 = pd.merge(data3, items, how='left')
data3 = pd.merge(data3, sales_sum, how='left')
data3 = pd.merge(data3, close_shop, how='left')
data3 = pd.merge(data3, mean_price, how='left')
data3 = pd.merge(data3, sales_lastmonth_by_itemshop, how='left')
data3 = data3.drop(['item_name','item_price'],axis=1)

data3['sales_last_month'].fillna(0, inplace=True)
data3['sales_sum'].fillna(0, inplace=True)
data3
```

```python
data3['sales_last_month'].sum()
```

# 建模

```python
train = data[:df_train_size]
test = data[df_train_size:]
train = train.drop_duplicates(['shop_id','item_id'])
X, y = data3.drop('sales_last_month',axis=1), data3['sales_last_month']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=666)
```

```python
X
```

```python
y
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor


def BoostingRegression(max_depth=4, n_estimators=100, min_samples_leaf=60, learning_rate=0.1, subsample=0.8):
    return Pipeline([
            ("gb_reg", GradientBoostingRegressor(max_depth=max_depth, n_estimators=n_estimators,
                                                                    learning_rate=learning_rate, min_samples_split=1200,
                                                                 subsample=subsample,
                                                                min_samples_leaf=min_samples_leaf))
    ] )
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = [
    {
        'gb_reg__max_depth': [2,5],
        'gb_reg__n_estimators': [30,400],
        'gb_reg__min_samples_leaf': [60],
        'gb_reg__learning_rate': [0.2,1],
        'gb_reg__subsample': [0.8]
    }
]

gb_reg = BoostingRegression()
grid_search = GridSearchCV(gb_reg, param_grid, verbose=2, cv=3) # cv是交叉验证分成多少分
grid_search.fit(X_train, y_train)
```

```python
gboost_reg = grid_search.best_estimator_
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
gboost_reg.score(X_test, y_test)
```
