```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plot_x = np.linspace(-1, 6, 141)
plot_x
```

```python
plot_y = (plot_x - 2.5)**2-1
```

```python
plt.plot(plot_x, plot_y)
plt.show()
```

```python
def dJ(theta):    #求导
    return 2*(theta - 2.5) #针对该函数的导数
```

```python
def J(theta):
    return (theta-2.5)**2-1
```

```python
eta = 0.1
epsilon = 1e-8
theta = 0.0
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    
    if(abs(J(theta) - J(last_theta)) < epsilon):   #浮点数判断 == 0
        break
        
print(theta)
print(J(theta))
```

```python
theta = 0.0
theta_history = []
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    theta_history.append(theta)
    
    if(abs(J(theta) - J(last_theta)) < epsilon):   #浮点数判断 == 0
        break
        
plt.plot(plot_x, J(plot_x))
plt.plot(np.array(theta_history), J(np.array(theta_history)), color = 'r', marker = '+')
```

```python
def dJ(theta):    #求导
    return 2*(theta - 2.5) #针对该函数的导数

def J(theta):
    return (theta-2.5)**2-1

def gredient_descent(initial_theta, eta, epsilon=1e-8, n_iters = 1e4):
    theta = initial_theta
    theta_history.append(initial_theta)
    i_iter = 0
    
    while i_iter < n_iters:      ### i_iter 执行次数，n_iter 次数上限
        gradient = dJ(theta)
        last_theta = theta
        theta = theta - eta * gradient
        theta_history.append(theta)
        
        if(abs(J(theta) - J(last_theta)) < epsilon):   #浮点数判断 == 0
            break
        
        i_iter += 1
        return 
    
def plot_theta_history():
    plt.plot(plot_x, J(plot_x))
    plt.plot(np.array(theta_history), J(np.array(theta_history)), color = 'r', marker = '+')
    plt.show()
```

### eta 是一个超参数，需要网格搜索，一般默认 0.01
