### 简单线性回归

```python
#封装
import numpy as np
from sklearn.metrics import r2_score

class SimpleLinearRegression2:
    def _init_(self):
        """初始化Simple Linear 模型"""
        self.a_ = None
        self.b_ = None
    
    def fit(self, x_train, y_train):
        """根据训练数据集x_train, y_train训练Simple Linear Regression模型"""
        assert x_train.ndim == 1, \
            "Simple Linear Regression can only solve single feature training data."
        assert len(x_train) == len(y_train), \
            "the size of x_train must be equal to the size of y_train"
        
        x_mean = np.mean(x_train)
        y_mean = np.mean(y_train)
        
        num = (x_train - x_mean).dot(y_train - y_mean)   #利用向量计算
        d = (x_train - x_mean).dot(x_train - x_mean)
        
        self.a_ = num / d
        self.b_ = y_mean - self.a_ * x_mean
        
        return self
    
    def predict(self, x_predict):
        """给定待预测数据集x_predict, 返回表示x_predict的结果向量"""
        assert x_predict.ndim == 1, \
            "Simple Linear Regressor can only solve single feature training data"
        assert self.a_ is not None and self.b_ is not None, \
            "must fit before predict !"
        
        return np.array([self._predict(x) for x in x_predict])
    
    def _predict(self, x_single):
        """给定单个待预测数据x_single, 返回x_single的预测结果值"""
        return self.a_ * x_single + self.b_
    
    def score(self, x_test, y_test):
        """根据测试数据集 x_Test, y_Test 确定当前数据集的准确度"""
        
        y_predict = self.predict(x_test)
        return r2_score(x_test, y_predict)
    
    def __repr__(self):
        return "SimpleLinearRegression2()"
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2) #random_state=666 种子
```

```python
reg = SimpleLinearRegression2()
reg.fit(x_train, y_train)
```

```python
reg.a_
```

```python
reg.b_
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, reg.predict(x_train), color="r")
plt.show()
```

```python
y_predict = reg.predict(x_test)
```

### MSE

```python
mse_test = np.sum((y_predict - y_test)**2) / len(y_test)
mse_test
```

### RMSE

```python
from math import sqrt
rmse_test = sqrt(mse_test)
rmse_test
```

### MAE

```python
mae_test = np.sum(np.absolute(y_predict - y_test)) / len(y_test)
mae_test
```

### scikit-learn中的MSE和MAE

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
mean_squared_error(y_test, y_predict)
```

```python
mean_absolute_error(y_test, y_predict)
```

### 总结MSE、RMSE和MAE
MSE、RMSE因为有平方，所以具有放大最大误差项的作用

# 最好的衡量线性回归法的指标:R Squared

### 理解
y_mean - yi是基准模型的误差，y_hat - yi是我们自己模型得到误差（y_hat是我们自己的函数计算出来的）
所以其比值再被1减去后代表，我们的模型误差比y_mean优化的比率。

### 实际应用的计算过程

```python
#自己计算
from sklearn.metrics import mean_squared_error  #用mse来计算
import numpy as np

1 - mean_squared_error(y_test, y_predict) / np.var(y_test)
```

```python
#sklearn 中的方式
from sklearn.metrics import r2_score

r2_score(y_test, y_predict)
```
