```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1., 2., 3., 4., 5.])
y = np.array([1., 3., 2., 3., 5.])
```

```python
plt.scatter(x, y)
plt.axis([0, 6, 0, 6])
plt.show()
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
```

```python
num = 0.0
d = 0.0
for x_i, y_i in zip(x, y):
    num += (x_i - x_mean) * (y_i - y_mean)
    d += (x_i - x_mean) ** 2
```

```python
a = num / d
b = y_mean - a * x_mean
```

```python
a
```

```python
b
```

```python
y_hat = a * x + b
```

```python
plt.scatter(x, y)
plt.plot(x, y_hat, color='r') #再加一条直线
plt.axis([0, 6, 0, 6])
plt.show()
```

```python
#应用

x_predict = 6
y_predict = a * x_predict + b
```

### 封装

```python
import numpy as np

class SimpleLinearRegression1:
    def _init_(self):
        """初始化Simple Linear 模型"""
        self.a_ = None
        self.b_ = None
    
    def fit(self, x_train, y_train):
        """根据训练数据集x_train, y_train训练Simple Linear Regression模型"""
        assert x_train.ndim == 1, \
            "Simple Linear Regression can only solve single feature training data."
        assert len(x_train) == len(y_train), \
            "the size of x_train must be equal to the size of y_train"
        
        x_mean = np.mean(x_train)
        y_mean = np.mean(y_train)
        
        num = 0.0
        d = 0.0
        for x, y in zip(x_train, y_train):                  #zip函数将元素打包成元组
            num += (x - x_mean) * (y - y_mean)
            d += (x - x_mean) ** 2        
        self.a_ = num / d
        self.b_ = y_mean - self.a_ * x_mean
        
        return self
    
    def predict(self, x_predict):
        """给定待预测数据集x_predict, 返回表示x_predict的结果向量"""
        assert x_predict.ndim == 1, \
            "Simple Linear Regressor can only solve single feature training data"
        assert self.a_ is not None and self.b_ is not None, \
            "must fit before predict !"
        
        return np.array([self._predict(x) for x in x_predict])
    
    def _predict(self, x_single):
        """给定单个待预测数据x_single, 返回x_single的预测结果值"""
        return self.a_ * x_single + self.b_
    
    def __repr__(self):
        return "SimpleLinearRegression1()"
```

```python
#应用
reg1 = SimpleLinearRegression1()
reg1.fit(x, y)
```

```python
reg1.predict(np.array([x_predict]))
```

```python
reg1.a_
```

```python
reg1.b_
```

```python
y_hat1 = reg1.predict(x)
```

```python
plt.scatter(x, y)
plt.plot(x, y_hat1, color='r') #再加一条直线
plt.axis([0, 6, 0, 6])
plt.show()
```

### 向量化计算 （向量计算是预处理号的函数，可以加快速度）

```python
#封装

import numpy as np

class SimpleLinearRegression2:
    def _init_(self):
        """初始化Simple Linear 模型"""
        self.a_ = None
        self.b_ = None
    
    def fit(self, x_train, y_train):
        """根据训练数据集x_train, y_train训练Simple Linear Regression模型"""
        assert x_train.ndim == 1, \
            "Simple Linear Regression can only solve single feature training data."
        assert len(x_train) == len(y_train), \
            "the size of x_train must be equal to the size of y_train"
        
        x_mean = np.mean(x_train)
        y_mean = np.mean(y_train)
        
        num = (x_train - x_mean).dot(y_train - y_mean)   #利用向量计算
        d = (x_train - x_mean).dot(x_train - x_mean)
        
        self.a_ = num / d
        self.b_ = y_mean - self.a_ * x_mean
        
        return self
    
    def predict(self, x_predict):
        """给定待预测数据集x_predict, 返回表示x_predict的结果向量"""
        assert x_predict.ndim == 1, \
            "Simple Linear Regressor can only solve single feature training data"
        assert self.a_ is not None and self.b_ is not None, \
            "must fit before predict !"
        
        return np.array([self._predict(x) for x in x_predict])
    
    def _predict(self, x_single):
        """给定单个待预测数据x_single, 返回x_single的预测结果值"""
        return self.a_ * x_single + self.b_
    
    def __repr__(self):
        return "SimpleLinearRegression2()"
```

```python
#应用

reg2 = SimpleLinearRegression2()
reg2.fit(x, y)
```

```python
reg2.a_
```

```python
reg2.b_
```

```python
y_hat2 = reg2.predict(x)

plt.scatter(x, y)
plt.plot(x, y_hat2, color='r') #再加一条直线
plt.axis([0, 6, 0, 6])
plt.show()
```
