# 评价精准率

### 股票预测比较在意预测能涨的股票是否准确(精准率)，否则会导致投资错误，导致亏本。
### 股票预测不在意真实能涨的股票中我们猜对了多少(召回率)，因为没猜中也不会导致亏本

# 评价召回率

### 病人诊断比较在意所有真实患病的病人中我们预测出来了几个(召回率)
### 至于预测为患病的病人中有多少是精确的(精准率)，因为就算精准率低，也就是没患病的被预测为患病了，也可以让人来二次检查，但这样没法检测出更多的真实患病病人

### 实现

```python
import numpy as np
from sklearn import datasets
```

```python
digits = datasets.load_digits()
X = digits.data
y = digits.target.copy()   #这里的等号只是引用!!!

y[digits.target==9] = 1
y[digits.target!=9] = 0
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
log_reg.score(X_test, y_test)
```

```python
y_log_predict = log_reg.predict(X_test)
```

```python
def TN(y_true, y_predict):
    assert len(y_true) == len(y_predict)
    return np.sum((y_true == 0) & (y_predict == 0))

TN(y_test, y_log_predict)
```

```python
def FP(y_true, y_predict):
    assert len(y_true) == len(y_predict)
    return np.sum((y_true == 0) & (y_predict == 1))

FP(y_test, y_log_predict)
```

```python
def FN(y_true, y_predict):
    assert len(y_true) == len(y_predict)
    return np.sum((y_true == 1) & (y_predict == 0))

FN(y_test, y_log_predict)
```

```python
def TP(y_true, y_predict):
    assert len(y_true) == len(y_predict)
    return np.sum((y_true == 1) & (y_predict == 1))

TP(y_test, y_log_predict)
```

```python
def confusion_matrix(y_true, y_predict):
    return np.array([
        [TN(y_true, y_predict), FP(y_true, y_predict)],
        [FN(y_true, y_predict), TP(y_true, y_predict)]
    ])

confusion_matrix(y_test, y_log_predict)
```

```python
def precision_score(y_true, y_predict):
    tp = TP(y_true, y_predict)
    fp = FP(y_true, y_predict)
    try:
        return tp / (tp + fp)
    except:
        return 0.0
    
precision_score(y_test, y_log_predict)
```

```python
def recall_score(y_true, y_predict):
    tp = TP(y_true, y_predict)
    fn = FN(y_true, y_predict)
    try:
        return tp / (tp + fn)
    except:
        return 0.0
    
recall_score(y_test, y_log_predict)
```

### scikit-learn中的混淆矩阵，精准率和召回率

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_log_predict)
```

```python
from sklearn.metrics import precision_score

precision_score(y_test, y_log_predict)
```

```python
from sklearn.metrics import recall_score

recall_score(y_test, y_log_predict)
```
