```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10,100)
y = np.sin(x)
print(x)
plt.plot(x,y)
plt.show() 
```

```python
cosy = np.cos(x)
siny = np.sin(x)
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--",label="cos(x)") #linestyle( :  -.  --  - )
plt.xlim(-5,15) # 指定x轴范围
plt.ylim(0,1.5) # 指定y轴范围
plt.axis([-5,15,0,1.5]) # x轴，y轴范围，同上两句意义一样
plt.xlabel("x value") # 不支持中文
plt.ylabel("y value")
plt.title("this is title")

plt.legend() # 加上示例图,即plot里的label
plt.show() # 同时渲染两条
```

```python
## 散点图
plt.scatter(x,siny)
plt.scatter(x,cosy,color="red",alpha=0.5) #alpha 透明度
plt.show()
```
