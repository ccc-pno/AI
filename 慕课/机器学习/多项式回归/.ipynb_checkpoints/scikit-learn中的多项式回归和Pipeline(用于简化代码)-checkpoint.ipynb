```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-3, 3, size = 100)
X = x.reshape(-1, 1)
y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, size=100)
```

```python
from sklearn.preprocessing import PolynomialFeatures  # preprocessing 预处理？
```

```python
poly = PolynomialFeatures(degree=2)  # degree 代表 最高增加几次幂
poly.fit(X)
X2 = poly.transform(X)
```

```python
X.shape
```

```python
X2.shape
```

```python
X2[:5, :]      # 第一列加入的是X0次方，所以变成了3列
```

```python
X[:5, :]
```

```python
from sklearn.linear_model import LinearRegression

lin_reg2 = LinearRegression()
lin_reg2.fit(X2, y)
y_predict2 = lin_reg2.predict(X2)
```

```python
plt.scatter(x, y)
plt.plot(np.sort(x), y_predict2[np.argsort(x)], color='r')
plt.show()
```

```python
lin_reg2.coef_   #第一列X0次方系数是0
```

```python
lin_reg2.intercept_
```

### Pipline

```python
x = np.random.uniform(-3, 3, size = 100)
X = x.reshape(-1, 1)
y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, size=100)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

poly_reg = Pipeline([
        ("poly", PolynomialFeatures(degree=2)),
        ("std_scalert", StandardScaler())
] )
```

```python
poly_reg.fit(X,y)
```

```python
y_predict = poly_reg.predict(X)  # 不加带有fit函数的类，predict会出错
```

```python
### 正确应用

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures 

poly_reg = Pipeline([
        ("poly", PolynomialFeatures(degree=2)),
        ("std_scalert", StandardScaler()),
        ("lin_reg", LinearRegression())
] )
```

```python
poly_reg.fit(X, y)
```

```python
poly_reg.fit(X, y)
y_predict = poly_reg.predict(X)
```

```python
plt.scatter(x, y)
plt.plot(np.sort(x), y_predict[np.argsort(x)], color='r')
plt.show()
```
