# 子模型的优点

# Bagging 和 Pasting

# 放回取样是随机的，会造成某些样本取不到
# 可以将这些样本作为测试集合

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets

X, y = datasets.make_moons(n_samples=500, noise=0.3, random_state=42)
```

```python
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42) 
```

# 使用Bagging

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier

# n_estimators 多少个模型，max_samples 每个模型多少样本点
# bootstrap=True 放回取样，bootstrap=False不放回取样等同于Pasting

bagging_clf = BaggingClassifier(DecisionTreeClassifier(), 
                                               n_estimators=50, max_samples=100,
                                               bootstrap=True)
```

```python
bagging_clf.fit(X_train, y_train)
bagging_clf.score(X_test, y_test)
```

```python
bagging_clf2 = BaggingClassifier(DecisionTreeClassifier(), 
                                               n_estimators=500, max_samples=100,
                                               bootstrap=True)
bagging_clf2.fit(X_train, y_train)
bagging_clf2.score(X_test, y_test)
```

# 理论上来说 n_estimators 越大，score越高
# max_samples可以调参

# 放回取样是随机的，会造成某些样本取不到
# 可以将这些样本作为测试集合

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier


bagging_clf = BaggingClassifier(DecisionTreeClassifier(), 
                                               n_estimators=500, max_samples=100,
                                               bootstrap=True, oob_score=True)

bagging_clf.fit(X, y)   # 用了oob_score就不需要区分 X_train了！！！！！！！！
```

```python
bagging_clf.oob_score_
```

### 使用n_jobs+oob

```python
%%time
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier

bagging_clf = BaggingClassifier(DecisionTreeClassifier(), 
                                               n_estimators=500, max_samples=100,
                                               bootstrap=True, oob_score=True, 
                                                )

bagging_clf.fit(X, y)   # 用了oob_score就不需要区分 X_train了！！！！！！！！
```

```python
%%time
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier

bagging_clf = BaggingClassifier(DecisionTreeClassifier(), 
                                               n_estimators=500, max_samples=100,
                                               bootstrap=True, oob_score=True, 
                                                n_jobs=-1)

bagging_clf.fit(X, y)   # 用了oob_score就不需要区分 X_train了！！！！！！！！
```

# 对特征随机取样 bootstrap_features

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier

# 因为就两个特征，所以max_features=1
# bootstrap_features=True代表对特征放回取样
# max_samples=500, 因为就500个点，所以相当于没有对样本随机取样
random_subspaces_clf = BaggingClassifier(DecisionTreeClassifier(),
                               n_estimators=500, 
                                max_samples=500,  
                               bootstrap=True, oob_score=True,
                                n_jobs=-1,
                               max_features=1, bootstrap_features=True)
random_subspaces_clf.fit(X, y)
random_subspaces_clf.oob_score_
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier

# max_samples=100, max_features=1，既对样本随机取样，又对特征随机取样
random_patches_clf = BaggingClassifier(DecisionTreeClassifier(),
                               n_estimators=500, 
                               max_samples=100,
                               bootstrap=True, oob_score=True,
                               n_jobs=-1,
                               max_features=1, bootstrap_features=True)
random_patches_clf.fit(X, y)
random_patches_clf.oob_score_
```
