# x1-x

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def entropy(p):
    return -p * np.log(p) - (1-p) * np.log(1-p)
```

```python
x = np.linspace(0.01, 0.99, 200)
```

```python
plt.plot(x, entropy(x))
plt.show()
```
