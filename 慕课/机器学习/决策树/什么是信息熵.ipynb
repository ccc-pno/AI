# 信息熵公式(H>0)
##### {1/10,2/10,7/10}因为信息大部分都在第三类中，所以信息更稳定，更确定
##### {1，0，0}因为信息都在第一类中，所以信息稳定，完全确定

# 若信息只有两类，一类概率是x，另一类就是1-x

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def entropy(p):
    return -p * np.log(p) - (1-p) * np.log(1-p)
```

```python
x = np.linspace(0.01, 0.99, 200)
```

```python
plt.plot(x, entropy(x))
plt.show()
```

### 上图是   两类信息的信息熵函数
### 说明两信息都是0.5时，熵最大，最不稳定
