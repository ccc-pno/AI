# 基本用法

```python
x = torch.ones(2, 2,  requires_grad=True)
print(x)
```

```python
y = x+2
z = y*y+3
out = z.mean()

print(y)
print(z)
print(out)
```

```python
out.backward()
```

```python
print(x.grad)
```

```python
# 尝试无梯度
with torch.no_grad():
    k = x+1
    print(k)
```
